```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from matplotlib.pyplot import tight_layout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import learning_curve
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
sns.set()
```

### Reading data

```python
colnames=['timestamp','ch0_raw','ch1_raw','ch0_hp','ch1_hp','quat_x','quat_y','quat_z','quat_w','gyro_x','gyro_y','gyro_z','acc_x','acc_y','acc_z','label','repetition'] 
df = pd.read_csv('pison_data_interview.csv', names=colnames, header=None)
df.head(6)
```

### Calculating features
A sliding windowing technique on raw time series data within each motion was applied. The window size is 300 samples (900ms) with an overlap of 200 samples (600ms). Features such as mean, std, min, max for both time and frequency domains were calculated within the window.

```python
# Time window application on raw time series data
window_size = 300
step_size = 100

labels = []
segments = []
for lab in range(5):
    data = df[df.label == lab]
    for i in range(0, data.shape[0] - window_size, step_size):
        b = data.values[i: i + window_size, 1:15]
        c = stats.mode(data['repetition'][i: i + window_size])
        lab = c[0][0]
        segments.append(b)
        labels.append(lab)
```

```python
# Feature calculation within each time window
X = np.zeros((len(segments), 8*14))
for i in range(len(segments)):    
    temp = segments[i]
    MEAN = np.mean(temp, axis=0)
    STD = np.std(temp, axis=0)
    MIN = np.min(temp, axis=0)
    MAX = np.max(temp, axis=0)
    #SKEW = stats.skew(temp)
    #ENERGY = np.sum(temp**2, axis=0)/window_size
    
    temp_freq = abs(np.fft.fft(temp, axis=0))[0 : window_size//2]
    FMEAN = np.mean(temp_freq, axis=0)
    FSTD = np.std(temp_freq, axis=0)
    FMIN = np.min(temp_freq, axis=0)
    FMAX = np.max(temp_freq, axis=0)
    #FSKEW = stats.skew(temp_freq)
    #FENERGY = np.sum(temp_freq**2, axis=0)/window_size

    #X[i, :] = np.append(MEAN, [STD, MIN, MAX, SKEW, ENERGY, FMEAN, FSTD, FMIN, FMAX, FSKEW, FENERGY])
    #X[i, :] = np.append(MEAN, [STD, MIN, MAX, ENERGY, FMEAN, FSTD, FMIN, FMAX, FENERGY])
    X[i, :] = np.append(MEAN, [STD, MIN, MAX, FMEAN, FSTD, FMIN, FMAX])
```

### Classfier construction

```python
# Splitting data into training(80%) and testing(20%) datasets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size = 0.2, random_state = 42)

# Scaling data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# Machine learning model initialization and applying cross validation
estimator = GaussianNB()
train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(
    estimator,
    X_train_scaled,
    y_train,
    scoring='accuracy',
    cv=5,
    return_times=True,
)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# Plot learning curve
plt.figure(figsize=(8, 5))
plt.grid()
plt.fill_between(
    train_sizes,
    train_scores_mean - train_scores_std,
    train_scores_mean + train_scores_std,
    alpha=0.1,
    color="r",
)
plt.fill_between(
    train_sizes,
    test_scores_mean - test_scores_std,
    test_scores_mean + test_scores_std,
    alpha=0.1,
    color="g",
)
plt.plot(train_sizes, train_scores_mean, "o-", color="r", label="Training score")
plt.plot(train_sizes, test_scores_mean, "o-", color="g", label="Cross-validation score")

plt.title("Gaussian Naive Bayes Learning Curve", fontsize=18)
plt.xlabel("Training examples", fontsize=15)
plt.ylabel("Score", fontsize=15)
plt.legend(loc="right")
```

### Testing accuracy and prediction results

```python
# Fitting the data
estimator.fit(X_train_scaled, y_train)
y_pred = estimator.predict(X_test_scaled)
print("Testing accuracy:", accuracy_score(y_test, y_pred)*100, '%')
```

```python
# Plotting the confusion matrix
labels = ['1','2','3']
cm = confusion_matrix(y_pred,y_test)
sns.heatmap(cm, xticklabels=labels, yticklabels=labels, annot=True,linewidths = 0.1, fmt='d', cmap = 'YlGnBu')
plt.title('Confusion matrix', fontsize = 15)
plt.ylabel('true label')
plt.xlabel('predicted label')
plt.show()
```
